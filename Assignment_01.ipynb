### Case Study: Ares Manufacturing - Project Griffin

#### Case Background

Ares Manufacturing, a publicly traded leader in advanced materials, is considering a major strategic investment into a new product line, codenamed "Project Griffin." The project involves developing and manufacturing a new lightweight, high-strength composite material for the aerospace and automotive industries.

The management team has tasked you, a senior financial analyst, with conducting a comprehensive financial evaluation of Project Griffin to determine its viability and its potential impact on the company. You must also address related financial management questions concerning the company's cost of capital, portfolio investments, and risk management strategies.

The evaluation is divided into four parts. Please address all questions thoroughly, showing your calculations and explaining your reasoning. Adhere to four decimal places for intermediate calculations and final answers.

**General Assumptions:** For the purposes of this case, assume perfect capital markets. This means there are no transaction costs or bid-ask spreads unless explicitly stated. All securities are perfectly divisible and can be bought or sold at the quoted prices.

---

### Part 1: Determining the Firm's Cost of Capital (WACC)

Before evaluating the new project, you must first determine the appropriate WACC for Ares Manufacturing. The company's capital structure and market data are as follows:

**Equity Information:**
*   **Shares Outstanding:** 2,000,000
*   **Current Stock Price:** $50.00 per share
*   **Equity Beta (β_E):** 1.20

**Debt Information:**
*   Ares has one bond issue outstanding with a total face value of $40,000,000.
*   **Coupon Rate:** 6.0% (paid semi-annually)
*   **Years to Maturity:** 15 years
*   **Current Bond Price:** 98% of par value

**Market Information:**
*   **Risk-Free Rate (r_f):** 4.0%
*   **Market Risk Premium (E[R_M] - r_f):** 5.5%
*   **Corporate Tax Rate:** 35%

**Questions:**

1.  Calculate the market value of Ares's equity (E) and debt (D). Determine the capital structure weights (w_E and w_D).
2.  Calculate the cost of equity (r_E) using the Capital Asset Pricing Model (CAPM).
3.  Calculate the pre-tax cost of debt (r_D) by finding the Yield to Maturity (YTM) of the company's bonds.
4.  Calculate the Weighted Average Cost of Capital (WACC) for Ares Manufacturing.
5.  Discuss the appropriateness of using Ares Manufacturing's overall WACC as the discount rate for Project Griffin. What specific factors related to this new project might suggest a different discount rate should be used, and why?

### Part 2: Project Valuation (Capital Budgeting)

Now, evaluate Project Griffin using the WACC calculated in Part 1. The project is expected to have a 5-year life.

**Investment & Depreciation:**
*   **Initial Capital Expenditure (CapEx):** $7,500,000 at Year 0.
*   **Depreciation Method:** Straight-line over 5 years to a book value of zero.
*   **Salvage Value:** The equipment is expected to be sold for $1,000,000 at the end of Year 5.

**Revenue & Costs:**
*   **Sales Forecast (Units):**
    *   Year 1: 50,000
    *   Year 2: 55,000
    *   Year 3: 60,000
    *   Year 4: 50,000
    *   Year 5: 40,000
*   **Price per Unit:** $70.00.
*   **Variable Cost per Unit:** $28.00.
*   **Fixed Costs (SG&A):** $400,000 per year.

**Working Capital & Tax:**
*   **Net Working Capital (NWC):** The project requires an investment in NWC at the beginning of each year equal to 15% of that year's sales. The initial investment at Year 0 is based on Year 1 sales, and the full amount of invested NWC is recovered at the end of the project.
*   **Corporate Tax Rate:** 35%.

**Questions:**

1.  Create a pro-forma forecast of the incremental free cash flows (FCF) for Project Griffin for Years 0 through 5.
2.  Using the WACC calculated in Part 1, determine the Net Present Value (NPV) of the project.
3.  Calculate the project's Internal Rate of Return (IRR) and its Payback Period.
4.  Based on your analysis, should Ares Manufacturing proceed with Project Griffin? Justify your decision.
5.  Regardless of your initial NPV finding, management wants to understand the project's sensitivity to key assumptions. Identify two critical variables in your forecast (e.g., price, unit sales, variable cost). For each, calculate the break-even value that would make the project's NPV exactly zero. Discuss the business feasibility of achieving these break-even values.

### Part 3: Portfolio Management

The CFO of Ares Manufacturing is considering investing a portion of the company's excess cash into a portfolio of two stocks: Alpha Corp. and Beta Corp. The financial data for these stocks are as follows:

| Stock | Expected Return (E[R]) | Volatility (σ) |
|:---|:---|:---|
| Alpha Corp. | 15.0% | 25.0% |
| Beta Corp. | 9.0% | 18.0% |

*   **Correlation (ρ_AB):** The correlation coefficient between the returns of Alpha and Beta is 0.20.
*   **Risk-Free Rate (r_f):** 4.0%

**Questions:**

1.  Calculate the expected return and volatility of a portfolio consisting of 60% Alpha Corp. and 40% Beta Corp.
2.  Calculate the weights ($w_A$, $w_B$) for the Global Minimum Variance Portfolio (GMVP).
3.  Calculate the weights ($w_A$, $w_B$) for the Optimal Risky Portfolio (ORP or Tangency Portfolio).
4.  What are the expected return, volatility, and Sharpe Ratio of this Optimal Risky Portfolio?
5.  The Optimal Risky Portfolio (ORP) mathematically provides the highest Sharpe Ratio. However, the CFO is hesitant to invest solely based on this calculation. Discuss two potential real-world reasons why the CFO might choose a portfolio different from the ORP. Consider factors beyond expected return and volatility.

### Part 4: Derivatives and Risk Management

To manage risks associated with its suppliers, the CFO wants to evaluate derivatives related to a key publicly-traded supplier, **Zeta Minerals**.

**Scenario A: Futures Contract**
*   Ares enters into a **short** futures contract on Zeta Minerals stock.
*   **Contract Size:** 100 contracts (each contract is for 100 shares).
*   **Futures Price:** $75.00 per share.
*   **Initial Margin:** $25,000.
*   **Maintenance Margin:** $18,000.

**Scenario B: Options Pricing**
*   The current stock price of Zeta Minerals is $100 per share.
*   Over the next year, Zeta's stock price is expected to either increase by 20% or decrease by 20%.
*   The one-year risk-free rate is 4.0%.
*   A one-year European put option on Zeta's stock with a strike price of $105 is currently trading for $9.50.

**Questions:**

1.  For the futures contract in Scenario A, what is the stock price that would trigger a margin call for Ares?
2.  For the options in Scenario B, what should a one-year European call option with a strike price of $105 trade at if there are no arbitrage opportunities? (Use Put-Call Parity).
3.  Using the Binomial Model, calculate the price of a one-year European **call** option on Zeta Minerals' stock with a strike price of $100.
4.  Compare and contrast the risk management outcomes of using the short futures contract (Scenario A) versus using long put options to hedge a stock position. Discuss which instrument would be more appropriate if the CFO is primarily concerned with (a) locking in a precise selling price, versus (b) protecting against a price drop while retaining the potential for upside gains.
