### Further Adjustments to Free Cash Flow: Example

*   Alpine Gear Outfitters is considering introducing a new line of high-tech hiking boots. 
    * This three-year project requires an initial fixed asset investment (new manufacturing equipment) of $\$2.1$ million. The equipment will be depreciated straight-line to zero over its three-year tax life, after which it will be considered worthless for salvage purposes. 
    * The project is estimated to generate $\$1.5$ million in annual sales, with costs of $\$600,000$ per year. The applicable tax rate is $30$ percent, and the company's required return for this type of project is $11$ percent. 
    * What is the project's Net Present Value (NPV)?

    **Calculate the annual Operating Cash Flow (OCF):**
    
    $ \text{Annual Depreciation} = \frac{\text{Initial Investment}}{\text{Project Life}} = \frac{\$2,100,000}{3} = \$700,000 $
    
    $ OCF = (\text{Revenues} - \text{Costs}) \times (1 - \tau_c) + \tau_c \times \text{Depreciation} $
    
    $ OCF = (\$1,500,000 – \$600,000) \times (1 – 0.30) + 0.30 \times \$700,000 \quad \rightarrow \quad OCF = \$840,000 $

    **Calculate the project NPV:**
    
    $ NPV = \sum_{t=0}^{n} \frac{FCF_t}{(1+r)^t} = FCF_0 + \frac{FCF_1}{(1 + r)^1} + \dots + \frac{FCF_n}{(1 + r)^n} $
    
    $ NPV = \text{-Initial Investment} + \sum_{t=1}^{3} \frac{OCF}{(1+r)^t} $
    
    $ NPV = -\$2,100,000 + \frac{\$840,000}{(1 + 0.11)^1} + \frac{\$840,000}{(1 + 0.11)^2} + \frac{\$840,000}{(1 + 0.11)^3} \quad \text{or} \quad NPV = -\$2,100,000 + \$840,000 \times \left[ \frac{1 - \frac{1}{(1 + 0.11)^3}}{0.11} \right] = -\$47,283.60 $

### Capital Budgeting

**Capital Budget**
*   Lists the investments that a company plans to undertake.

**Capital Budgeting**
*   Process used to analyze alternative investments.
*   Ultimate goal is to determine the effect of the decision on the firm's cash flows.

**Incremental Earnings**
*   The amount by which the firm's earnings are expected to change as a result of the investment decision.

**Capital Budgeting Process**
*   Revenue and cost estimates
*   Incremental earnings forecast
*   Free cash flow and NPV
*   Choose among alternatives

### Capital Budgeting: Discount cash flows, not profits

*   Accounting earnings are often different from actual cash flows! Always discount **cash flows**, not accounting profits.

*   It costs **\$5,000** upfront (Year 0) and is expected to generate cash revenues of **\$3,500** in Year 1 and **\$2,500** in Year 2. The accountant depreciates the initial development cost straight-line over the two years.

    |                     | Year 1   | Year 2   |
    | :------------------ | :------- | :------- |
    | Cash inflow         | +$3,500  | +$2,500  |
    | Less depreciation   | - 2,500  | - 2,500  |
    | Accounting Income | +$1,000| $0   |

    *   If we mistakenly discount *accounting income* at a 12% cost of capital:
    $$ NPV = \frac{\$1,000}{1.12^1} + \frac{\$0}{1.12^2} = \$892.86 \ \ \ \mathbf{<-- WRONG!} $$

    *   Instead, we must discount the actual *cash flows*:
    $$ NPV = -\$5,000 + \frac{\$3,500}{1.12^1} + \frac{\$2,500}{1.12^2} $$
    $$ NPV = -\$5,000 + \$3,125.00 + \$1,992.98 = \mathbf{\$117.98 \ \ \ <-- CORRECT!} $$

### Capital Budgeting: Working capital matters

*   Changes in working capital affect the *timing* of cash flows and can differ significantly from when income is recognized.

*   A consulting project requires paying specialist fees of **\$8,000** at (t=1). The client is billed **\$15,000** upon project completion (t=1) and pays at t=2.

    **Accounting perspective (assuming income recognized when billed):**
    |                     | Amount(\$)  | Timing |
    | :------------------ | ------: | -----: |
    | Revenue             | 15,000 | t=1    |
    | Less cost           | - 8,000 | t=1    |
    | Accounting Income | $ 7,000 | t=1|

    **Cash Flow perspective:**
    |                            | t=0 | t=1 | t=2 |
    | :------------------------- | -------: | -------: | -------: |
    | **Components of CF:**      |          |          |          |
    | Cash Outlay (Specialists)  |      0   | -8,000   |       0  |
    | Cash Inflow (Client Pmt)   |      0   |      0   | +15,000  |
    | Net Cash flow              |      0   | -8,000   | +15,000  |
       

* The timing difference between cash flow and accounting income can significantly impact project valuation. 
* **Always track the actual cash inflows and outflows.**

### Capital Budgeting: Discount incremental cash flows

*   The value of a project depends on all the additional cash flows that follow from project acceptance.
*   Include all incidental effects
    *   Introduction of a new product will almost certainly cut into sales of existing products (Product Cannibalization).
*   Do not confuse average with incremental payoffs
    *   The incremental NPV from investing in a loser may be strongly positive.
    *   Also, it does not always make sense to throw good money after good.
*   Forecast product sales but also recognize after-sales cash flows.
*   Include opportunity costs.
*   Forget sunk costs.
    *   They are past and irreversible outflows.
*   Beware of allocated overhead costs.
    *   In investment appraisal we should include only the extra expenses that would result from the project.
*   Remember salvage value
    *   When the project comes to an end, you may be able to sell the plant and equipment or redeploy the assets elsewhere in the business.

### Capital Budgeting: Discount incremental cash flows

*   GreenSpark Energy, a manufacturer of residential solar panels, is considering introducing a next-generation, higher-efficiency panel. The firm anticipates that annual sales of the new panel will be $50 million. While many sales will attract new customers, GreenSpark estimates that 30% will come from existing customers who opt for the new panel instead of the current model.

    *   Assume customers spend roughly the same amount on either panel model. What level of *incremental sales* is associated with introducing the new high-efficiency panel?

        *Incremental Sales = Total New Product Sales * (1 - Cannibalization Rate)*

        *Incremental Sales = $\$50$ million * (1 – 30%) = $\$50$ million * 0.70 = $\$35$ million*
    

    *   Now, suppose GreenSpark's market research indicates that 60% of the customers who would potentially switch from the original panel to the new one *would likely switch to a competitor's high-efficiency panel* if GreenSpark doesn't launch its own new model. What is the level of incremental sales in this case?

    *Incremental Sales = Total Sales * (1 - Cannibalization Rate + Lost Sales Avoided)*
    
    $\qquad \qquad \qquad \; \; \; = \$50m * (1 - 30\% + 18\%) = \$50m * 88\% = \$44 \text{million}$
        
    *Incremental Sales = Total New Product Sales * (1 – (Original Cannibalization Rate * % Staying Customers))*
 
        
    *Incremental Sales = $\$50m$ * [(1 – (30% * (1 - 60%))] = $\$44$ million*

### Capital Budgeting: Discount incremental cash flows

AeroDrone Dynamics currently sells its popular "SkyHawk" surveillance drone for $\$1,500$. It is considering lowering the price to $\$1,300$ next year. The cost of goods sold for the SkyHawk is $\$800$ per unit. Current annual sales are 5,000 units.

*   Suppose that if AeroDrone drops the price to $\$1,300$ *immediately* (this year), it can increase this year's sales volume by 30% to 6,500 units. What would be the incremental impact on this year's EBIT from this immediate price drop?


    $EBIT = \text{Sales Volume} * (\text{Price - COGS)}$
        
    $EBIT_{New} = 6,500\, \text{units} * (\$1,300 - \$800) = 6,500 * \$500 = \$3,250,000$
    
    $EBIT_{Old} = 5,000\, \text{units} * (\$1,500 - \$800) = 5,000 * \$700 = \$3,500,000$
        
    $\text{Change in EBIT} = EBIT_{New} - EBIT_{Old}$
    
    $\text{Change in EBIT} = \$3,250,000 - \$3,500,000 = -\$250,000$

### Capital Budgeting: Treat inflation consistently

**Discount nominal cash flows using nominal discount rate, real cash flows using the real discount rate.**

Suppose your software company typically forecasts project cash flows in nominal terms and discounts them at a $12\%$ nominal cost of capital. However, for a new app development project, the initial estimates are provided in today's dollars (real terms), and inflation is projected to be $5\%$ per year.

| Real Cash Flows (\$ thousands) | $C_0$ | $C_1$ | $C_2$ | $C_3$ |
| :---------------------------- | :---: | :---: | :---: | :---: |
| **Value**                     | -150  | +60   | +80   | +50   |

$\text{Real discount rate} (r_r) = \frac{1 + \text{Nominal discount rate}}{1 + \text{Inflation rate}} - 1 \quad \rightarrow \quad r_r = \frac{1 + 0.12}{1 + 0.05} - 1 = 6.67\% $

**Two Approaches to Calculate NPV:**

1.  Discount Real Cash Flows at the Real Discount Rate (6.67%)
2.  Convert Real Cash Flows to Nominal, then Discount at the Nominal Discount Rate (12.00%)

| t   | Real CF (\$k) | Nominal CF (\$k) [Real * (1+e)^t] | PV Real CF @ 6.67% (\$k) | PV Nominal CF @ 12.00% (\$k) |
| :-: | -----------: | -----------------------------: | ----------------------: | ------------------------: |
| 0   | -150.00       | -150.00                         | -150.00                  | -150.00                    |
| 1   | 60.00         | 63.00                           | 56.25                    | 56.25                      |
| 2   | 80.00         | 88.20                           | 70.31                    | 70.31                      |
| 3   | 50.00         | 57.88                           | 41.18                    | 41.18                      |
|     |               |                                 | NPV = 17.74          | NPV = 17.74            |

### Capital Budgeting

#### Separate investment and financing decisions
*   **Separation Principle:** When securities are fairly priced, the NPV of financing is zero, and so the NPV of an investment is independent of how it is financed.

#### Remember to Deduct Taxes
*   Taxes are an expense just like wages and raw materials.
*   Therefore, cash flows should be estimated on an after-tax basis.
*   Subtract cash outflows for taxes from pretax cash flows and discount the net amount.

### Determining Free Cash Flow and NPV
*   Earnings are just an accounting measure of the firm's performance.
    *   Earnings include non-cash charges, such as depreciation, but do not include the CAPEX.
*   The incremental effect of a project on the firm's available cash, separate from any financing decisions, is the project's free cash flow.
*   FCF calculation starts with Unlevered Net Income (UNI) and adjusts for non-cash expenses and cash flows related to investments.
    *   Add back non-cash expenses (like depreciation).
    *   Subtract capital expenditures (CAPEX).
    *   Subtract "increases in net working capital (NWC) or add decreases in NWC.

### Determining Free Cash Flow and NPV
*   Most projects require investments in Net Working Capital (NWC) to support operations. 
    * NWC represents the difference between short-term assets and short-term liabilities needed for the project.
    
    $ \text{Net Working Capital} = \text{Current Assets} - \text{Current Liabilities} \quad \rightarrow \quad NWC = (\text{Cash}_\text{req} + \text{Inventory} + \text{Receivables}) - (\text{Payables})$

    *   Accounts Payable are projected to be **20%** of the annual Cost of Goods Sold (COGS).

### Calculating Free Cash Flow Directly

1.  **Using Unlevered Net Income:**
    $$ \text{Free Cash Flow} = \overbrace{(\text{Revenues} – \text{Costs} - \text{Depreciation}) \times (1 – \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$
    *   Where "Costs" represent operating costs excluding depreciation (e.g., COGS + SG&A).
    *   $\tau_c$ is the corporate tax rate.

2.  **Using operating cash flow before depreciation tax shield:**
    $$ \text{Free Cash Flow} = (\text{Revenues} – \text{Costs}) \times (1 – \tau_c) – \text{CapEx} – \Delta \text{NWC} + (\tau_c \times \text{Depreciation}) $$
    *   This formula calculates the after-tax cash flow from operations as if there were no depreciation, then separately adds the tax savings provided by the depreciation deduction (the Depreciation Tax Shield, $\tau_c \times \text{Depreciation}$).

3.  **Starting from EBIT:**
    $$ \text{Free Cash Flow} = \overbrace{\text{EBIT} \times (1 – \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$

### Present Value Calculation

$$ PV(FCF_t) = \frac{FCF_t}{(1+r)^t} = FCF_t \times \overbrace{\frac{1}{(1+r)^t}}^{\text{Discount Factor for Year t}} $$

### Calculating Free Cash Flow (FCF)

*   Suppose you manage a growing consulting firm and considering upgrading the office's technology infrastructure (servers, network gear). 
    * This upgrade is expected to increase annual billable revenues by $\$70,000$ due to improved efficiency and capabilities, but also increase annual operating expenses (support, licenses, power) by $\$40,000$. 
    * The depreciation expense associated with the new technology for the first year is estimated at $\$15,000$. 
    * Your firm's marginal tax rate is $30\%$. 
    * Minor capital expenditures of $\$5,000$ will be required next year for peripheral component replacements, and no change in net working capital is anticipated for this specific project. 
    * Calculate the project's Free Cash Flow (FCF) for the next year.

    $ \text{Free Cash Flow} = (\text{Revenues} – \text{Costs}) \times (1 – \tau_c) – \text{CapEx} – \Delta NWC + \tau_c \times \text{Depreciation} $

    $ \text{FCF} = (\$70,000 – \$40,000) \times (1 – 0.30) – \$5,000 - \$0 + 0.30 \times \$15,000 $
    
    $ \text{FCF} = (\$30,000) \times (0.70) – \$5,000 + \$4,500 $
    
    $ \text{FCF} = \$21,000 – \$5,000 + \$4,500 $
    
    $ \text{FCF} = \$20,500 $

### Calculating Operating Cash Flow (OCF)

*   A local bakery, estimates that its new line of gluten-free pastries will generate annual sales of $\$500,000$. The associated operating costs (ingredients, labor, packaging) are expected to be $\$250,000$ per year, and the annual depreciation expense on the new specialized ovens and mixers is $\$50,000$. If the bakery's tax rate is $25\%$, what is the expected annual Operating Cash Flow (OCF) from this new line?

    **Method 1: Tax Shield Approach**
    
    $ OCF = (\text{Revenues} – \text{Costs}) \times (1 - \tau_c) + (\tau_c \times \text{Depreciation}) $
    
    $ OCF = (\$500,000 – \$250,000) \times (1 – 0.25) + (0.25 \times \$50,000) $
    
    $ OCF = (\$250,000) \times (0.75) + \$12,500 \qquad \rightarrow \qquad OCF = \$200,000 $

    **Method 2: Unlevered Net Income + Depreciation**
    
    $ \text{Free Cash Flow} = \overbrace{(\text{Revenues} – \text{Costs} - \text{Depreciation}) \times (1 − \tau_c)}^{\text{Unlevered Net Income}} + \text{Depreciation} - \text{CapEx} - \Delta NWC $
    
    $ \text{Unlevered NI} = (\text{Revenues} – \text{Costs} - \text{Depreciation}) \times (1 − \tau_c) $
    
    $ \text{Unlevered NI} = (\$500,000 – \$250,000 – \$50,000) \times (1 – 0.25) \qquad \rightarrow \qquad  \text{Unlevered NI} = (\$200,000) \times (0.75) = \$150,000 $

    $ OCF = \text{Unlevered Net Income} + \text{Depreciation} - \text{CapEx} - \Delta NWC $
    
    $ OCF = \$150,000 + \$50,000 = \$200,000 $

### Break-even analysis – DiagnosTech

- **Initial investment (Year 0)**:  $\quad FCF_0 = -(R\&D \times (1 - Tax) + CapEx) $
- **Year 1 FCF**: $\qquad \qquad \qquad \quad FCF_1(Q, P, VC) = OCF_1(Q, P, VC) - NWC_1(Q, P, VC)$
- **Years 2–4 FCF**:  $\qquad \qquad \qquad FCF_{2-4}(Q, P, VC) = OCF(Q, P, VC)$
- **Year 5 FCF**: $\qquad \qquad \qquad \quad FCF_5(Q, P, VC) = OCF_5(Q, P, VC) + NWC_1(Q, P, VC)$

- **Operating Cash Flow (OCF)**: $\quad OCF(Q, P, VC) = \left[ (Q \cdot P - Q \cdot VC - SG\&A - Depreciation) \cdot (1 - Tax) \right] + Depreciation$

- **Net Working Capital (NWC)**: $\quad NWC_1(Q, P, VC) = 0.10 \cdot (P \cdot Q) - 0.20 \cdot (VC \cdot Q)$

**Objective Function for Break-Even Calculation:**

The break-even point occurs when the project's Net Present Value (NPV) is zero. Thus, we have the general form:

$NPV(Q,P,VC,r) = FCF_0 + \frac{FCF_1(Q,P,VC)}{(1+r)^1} + \frac{FCF_2(Q,P,VC)}{(1+r)^2} + \frac{FCF_3(Q,P,VC)}{(1+r)^3} + \frac{FCF_4(Q,P,VC)}{(1+r)^4} + \frac{FCF_5(Q,P,VC)}{(1+r)^5}=0$

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
price = 450.0                  # Wholesale price per unit ($)
cogs = 180.0                   # Cost of goods per unit ($)
fixed_SGA = 3500               # Annual SG&A ($000)
annual_depr = 2000             # Annual depreciation ($000)
tax_rate = 0.35                # Tax rate
capex_initial = 10000          # Initial CapEx at year 0 ($000)
RD_initial = 15500             # Initial R&D expense at year 0 ($000)
NWC_year1 = 450                # NWC outflow in Year 1 ($000)
NWC_year5 = -450               # NWC inflow in Year 5 ($000)
discount_rate = 0.12           # Discount rate (12%)

# Years array
years = np.arange(6)

# FCF calculation with immediate tax deduction on R&D
def calculate_fcfs(units):
    sales = units * price / 1000
    cost_of_goods = units * cogs / 1000

    # EBIT and Taxes from year 1 onwards
    ebit = sales - cost_of_goods - fixed_SGA - annual_depr
    tax = ebit * tax_rate
    net_income = ebit - tax

    fcfs = np.zeros(6)

    # Year 0 includes tax shield from immediate R&D expense
    fcfs[0] = -(RD_initial * (1 - tax_rate) + capex_initial)

    # Years 1 to 5 FCFs
    fcfs[1] = net_income + annual_depr - NWC_year1
    fcfs[2:5] = net_income + annual_depr
    fcfs[5] = net_income + annual_depr - NWC_year5

    return fcfs

# NPV calculation
def npv(fcfs, rate=discount_rate):
    return sum(fcfs[t] / (1 + rate)**t for t in range(len(fcfs)))

# Plot setup
units_range = np.linspace(10000, 70000, 200)
npv_values = [npv(calculate_fcfs(u)) for u in units_range]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(units_range, npv_values, label='NPV vs Units Sold')
plt.axhline(0, color='black', linewidth=0.7, linestyle='--')
plt.xlabel('Units Sold per Year')
plt.ylabel('NPV ($000)')
plt.title('NPV vs. Units Sold')
plt.grid(True)

# Break-even annotation
break_even_units = np.interp(0, npv_values, units_range)
plt.scatter(break_even_units, 0, color='red', label=f'Break-even: {break_even_units:.0f} Units')
plt.annotate(f'{break_even_units:.0f} Units', 
             (break_even_units, 0), 
             xytext=(break_even_units + 2000, max(npv_values)/10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Re-import required libraries and re-run sensitivity analysis

import numpy as np
import matplotlib.pyplot as plt

# Redefine constants (initial assumptions)
units_initial = 50000
price_initial = 450
cogs_initial = 180
discount_initial = 0.12

fixed_SGA = 3500
annual_depr = 2000
tax_rate = 0.35
capex_initial = 10000
RD_initial = 15500
NWC_year1 = 450
NWC_year5 = -450

# Define scenarios (Worst, Base, Best)
scenarios = {
    'Units Sold': [30000, units_initial, 75000],
    'Sale Price': [250, price_initial, 575],
    'Cost of Goods': [300, cogs_initial, 120],
    'Cost of Capital': [0.18, discount_initial, 0.10]
}

# FCF calculation with immediate R&D tax shield
def calculate_fcfs(units, price, cogs):
    sales = units * price / 1000
    cost_of_goods = units * cogs / 1000
    ebit = sales - cost_of_goods - fixed_SGA - annual_depr
    tax = ebit * tax_rate
    net_income = ebit - tax
    fcfs = np.zeros(6)
    fcfs[0] = -(RD_initial * (1 - tax_rate) + capex_initial)
    fcfs[1] = net_income + annual_depr - NWC_year1
    fcfs[2:5] = net_income + annual_depr
    fcfs[5] = net_income + annual_depr - NWC_year5
    return fcfs

# NPV calculation
def npv(rate, fcfs):
    return sum(fcfs[t] / (1 + rate)**t for t in range(len(fcfs)))

# Compute base NPV
base_npv = npv(discount_initial, calculate_fcfs(units_initial, price_initial, cogs_initial))

# Compute NPVs for each scenario
npv_ranges = {}
for param, (worst, base, best) in scenarios.items():
    if param == 'Units Sold':
        low = npv(discount_initial, calculate_fcfs(worst, price_initial, cogs_initial))
        high = npv(discount_initial, calculate_fcfs(best, price_initial, cogs_initial))
    elif param == 'Sale Price':
        low = npv(discount_initial, calculate_fcfs(units_initial, worst, cogs_initial))
        high = npv(discount_initial, calculate_fcfs(units_initial, best, cogs_initial))
    elif param == 'Cost of Goods':
        low = npv(discount_initial, calculate_fcfs(units_initial, price_initial, worst))
        high = npv(discount_initial, calculate_fcfs(units_initial, price_initial, best))
    elif param == 'Cost of Capital':
        low = npv(worst, calculate_fcfs(units_initial, price_initial, cogs_initial))
        high = npv(best, calculate_fcfs(units_initial, price_initial, cogs_initial))
    npv_ranges[param] = [low, high]

# Prepare plot
labels = list(npv_ranges.keys())
lows = [npv_ranges[label][0]/1000 for label in labels]
highs = [npv_ranges[label][1]/1000 for label in labels]
base_million = base_npv / 1000

# Plot tornado diagram
fig, ax = plt.subplots(figsize=(10, 6))
for i, label in enumerate(labels):
    low = lows[i]
    high = highs[i]
    ax.barh(label, high - low, left=low, color='green' if high > low else 'red', alpha=0.6)
    ax.axvline(base_million, color='black', linestyle='--')

ax.set_xlabel('Project NPV ($ Millions)')
ax.set_title('Sensitivity Analysis - Tornado Diagram')
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

npv_ranges, base_npv
```
